```python
import pandas as pd
```

```python
df = pd.read_csv('subjects_diseases.csv')
```

```python
kidney_diseases_df = df[df['long_title'].str.contains('kidney', case=False)]
```

```python
kidney_diseases_df
```

```python

medicine_df = pd.read_csv('PRESCRIPTIONS.csv')
medicine_df
```

```python
unique_subjects = set(kidney_diseases_df['subject_id'])

# Step 2: Filter the medicines dataset based on the unique subject_ids
filtered_medicines_df = medicine_df[medicine_df['subject_id'].isin(unique_subjects)]
```
